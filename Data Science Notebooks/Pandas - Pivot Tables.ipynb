```python
import pandas as pd
```

```python
df = pd.read_csv('data/vgsales.csv')
```

```python
df
```

```python
#Create a pivot table 

df.groupby(['Publisher', 'Genre'])['NA_Sales','EU_Sales'].sum()
```

```python
#Plot by platform

import matplotlib.pyplot as plt

df.groupby(['Platform'])['Global_Sales'].sum().sort_values().plot(kind='barh', figsize=(15,8))
plt.title('Sales by Platform')
plt.xlabel("Global Sales")
```
