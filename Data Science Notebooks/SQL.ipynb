```python
import sqlite3 
conn = sqlite3.connect('data/database 2.sqlite')
cur = conn.cursor()
```

```python
#Select data 

cur.execute("""SELECT * 
                FROM Umpire; 
                """)
```

```python
#Put into a Pandas DF

import pandas as pd

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df.head()
```

```python
#Select with conditions

cur.execute("""SELECT Player_Id, Player_Name 
                FROM Player
                WHERE Country_Name = 4; 
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df.head()
```

```python
#Between 

cur.execute("""SELECT *
                FROM Player
                WHERE Country_Name BETWEEN 2 AND 5; 
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df.head()
```

```python
#Order by

cur.execute("""SELECT * 
                FROM Player
                WHERE Country_Name = 4 or Country_Name = 5
                ORDER BY Bowling_skill DESC; 
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

```python
#Limit by (oldest)

cur.execute("""SELECT * 
                FROM Player
                ORDER BY DOB ASC
                LIMIT 1; 
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

```python
#Null

cur.execute("""SELECT * 
                FROM Player
                WHERE Bowling_skill IS null; 
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

```python
# Count

cur.execute("""SELECT COUNT (Player_Id) 
                FROM Player
                WHERE Bowling_skill = 4.0; 
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

```python
#Group By 

cur.execute("""SELECT Player.Country_Name,
                COUNT (Player_Name) AS Num_Players 
                FROM Player
                GROUP BY Country_Name; 
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

```python
#Group By multiple 

cur.execute("""SELECT Country_Name, Bowling_skill, 
                COUNT (Player_Name) AS Num_Players 
                FROM Player
                GROUP BY Country_Name, Bowling_skill; 
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

```python
#Having (after grouping)

cur.execute("""SELECT Country_Name, Bowling_skill, 
                COUNT (Player_Name) AS Num_Players 
                FROM Player
                GROUP BY Country_Name, Bowling_skill
                HAVING Bowling_skill > 10.0; 
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

## Join Statements

```python
cur.execute("""SELECT * 
               FROM Match
               JOIN Venue
               ON Match.Venue_Id = Venue.Venue_Id
               LIMIT 2;
               """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

```python
cur.execute("""SELECT * 
               FROM Match
               JOIN Venue
               USING (Venue_Id)
               LIMIT 2;
               """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

```python
#Left join + check for missing values 

cur.execute("""SELECT * 
               FROM Match
               JOIN Venue
               USING (Venue_Id)
               """)
df = pd.DataFrame(cur.fetchall()) 
df.columns = [x[0] for x in cur.description]

print(len(df))
print(len(df[df.Venue_Id.isnull()]))
df[df.Venue_Id.isnull()].head()
```

## Multiple Joins

```python
cur.execute("""SELECT * 
               FROM Venue
               JOIN City
               USING (City_Id)
               JOIN Country
               ON City.Country_id = Country.Country_Id
               WHERE Country_Name = 'South Africa'
               ORDER BY Country_Name
               """)
df = pd.DataFrame(cur.fetchall()) 
df.columns = [x[0] for x in cur.description]
df
```

## Sub Query

```python
cur.execute("""SELECT * 
               FROM Venue
               JOIN City
               USING (City_Id)
               WHERE Country_Id = 1
               """)

cur.execute("""SELECT * 
               FROM Venue
               WHERE City_Id IN (SELECT City_Id
                                 FROM City
                                 WHERE Country_Id = 1);
                                 """)
df = pd.DataFrame(cur.fetchall()) 
df.columns = [x[0] for x in cur.description]
df
```
