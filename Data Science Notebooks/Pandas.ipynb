## Slicing with Conditions

```python
#Select where values are greater than (year > 2010)

df.loc[df['Year'] > 2010]
```

```python
#Select by multiple conditions (After 2010 and Sports)

df.loc[(df['Year'] > 2010) & (df['Genre'] == 'Sports')]
```

```python
#Count on conditions

EA_Sports_After2010 = df.loc[(df['Year'] > 2010) & (df['Genre'] == 'Sports') & (df['Publisher'] == 'Electronic Arts')]
print(len(EA_Sports_After2010))
```

```python
#Count categories with conditions

#How many publishers made a game in 2015?

Games2015 = df[df['Year'] == 2015]
publishers2015 = Games2015['Publisher'].unique()
print(len(set(publishers2015)))
```

```python
#Select a column based on another columns value (NA_Sales where year > 2010)

df.loc[df['Year'] > 2010, ['NA_Sales']]
```

## Manipulating Data

```python
#Change all values that are greater than 20 to 20 

df.loc[df['Global_Sales'] > 20, 'Global_Sales'] = 20
```

```python
#Rename Nintendo to Pintendo

df.loc[df['Publisher'] == 'Nintendo', 'Publisher'] = 'Pintendo'
```

## Manipulating Rows

```python
#Drop the 1st row

df = df.drop(0)
```

## Manipulating Columns

```python
#Add a column

df['Projected_Sales'] = 0
```

```python
#Drop a column

df = df.drop('Other_Sales', axis=1) 
```

```python
#Merge columns 

df['EU_and_Japan_Sales'] = df['EU_Sales'] + df['JP_Sales'] 
```

```python
#Function if a letter is in a word (useful for codes)

def contains_p(text):
    bool_val = 'N' in text
    return bool_val

df['On_P'] = df['Platform'].map(contains_p)
```

```python
df
```

```python
# Add column based on condition (hit if north american sales are greater than 1)

df.loc[df['NA_Sales'] > 5, 'Hit'] = 1
df.loc[df['NA_Sales'] <= 5, 'Hit'] = 0
```
