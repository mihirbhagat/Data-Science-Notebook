```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/train_users_2.csv')
```

```python
df.head()
```

```python
#Change to date time

df['date_account_created'] = pd.to_datetime(df['date_account_created'])
df['timestamp_first_active'] = pd.to_datetime((df.timestamp_first_active // 1000000), format='%Y%m%d')
df['date_first_booking'] = pd.to_datetime(df['date_first_booking'])
```

```python
date_account_created = np.vstack(df.date_account_created.astype(str).apply(lambda x: list(map(int, x.split('-')))).values)
timestamp_first_active = np.vstack(df.timestamp_first_active.astype(str).apply(lambda x: list(map(int, x.split('-')))).values)
```

```python
#Separate date into year, month & day

df['created_year'] = date_account_created[:,0]
df['created_month'] = date_account_created[:,1]
df['created_day'] = date_account_created[:,2]
```

```python
df['lag'] = df['date_account_created'] - df['timestamp_first_active']
```

```python
df['lag']
```

```python
#Change order of dates

df['date_account_created'] = df['date_account_created'].dt.strftime("%m-%d-%y")
```

```python
#Find which day

df['weekday'] = df['timestamp_first_active'].dt.dayofweek
```
