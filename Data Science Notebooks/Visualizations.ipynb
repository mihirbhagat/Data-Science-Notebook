```python
#Imports

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Scatter Plot

```python
#Create dummy data

#30 data points between 0 and 10
x = np.linspace(0, 10, 30)
# Calcuate sin(x) and save in a new array y
y = np.linspace(0, 100, 30)
```

```python
#Set the figure size in inches
fig = plt.figure(figsize=(10,6))

plt.scatter(x, y, label = "X/Y Relationship" )

#Label X & Y
plt.xlabel('X Values')
plt.ylabel('Y Values')

#Label Title
plt.title('Scatter Plot in Matplotlib')

plt.legend()
plt.show()
```

## Sub Plots

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(10,10))

for i in range(1,13):
    row = (i-1)//4
    col = i%4-1
    ax = axes[row][col]
    y_new = [xi*i for xi in x]
    ax.plot(x, y_new)
    ax.set_title('{}*x'.format(i))
```

## Bar Graphs

```python
x = ['Crime', 'Sci/fi', 'Drama', 'Comedy', 'Action', 'Documentary' ]

y = [40, 30, 10, 50, 25, 5]
```

```python
plt.figure(figsize=(10,6))

#Plot bar graph 
plt.bar(x, y)

plt.title('Video Library')
plt.ylabel('Number of movies')
plt.xlabel('Genres')

plt.show()
```

## Histogram

```python
# Set seed for reproducability
np.random.seed(100)

# Generate a data set of 200 retirement age values
x = 5*np.random.randn(200) + 65
```

```python
plt.hist(x, bins = 10, edgecolor='black')

plt.xlabel('Retirement Age')
plt.ylabel('Frequency of Values')
plt.title('Histograms in Matplotlib')
plt.show()
```
